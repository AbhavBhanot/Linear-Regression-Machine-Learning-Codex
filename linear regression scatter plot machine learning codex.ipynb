```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv("/kaggle/input/salary-dataset-simple-linear-regression/Salary_dataset.csv")
```

```python
plt.scatter(df['YearsExperience'], df['Salary'])
plt.xlabel('x')
plt.ylabel('y')
plt.title("scatter plot")
plt.show()
```

```python
x = df[['YearsExperience']]
y = df['Salary']
```

```python
model = LinearRegression()
model.fit(x,y)
slope = model.coef_[0]
intercept = model.intercept_
```

```python
def linear_regression(x):
    return slope*x + intercept
```

```python

predicted_values = model.predict(df[['YearsExperience']])

plt.scatter(df['YearsExperience'], df['Salary'], label='Data Points')
plt.plot(df['YearsExperience'], predicted_values, color='red', label='Regression Line')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.title('Salary vs. Years of Experience')
plt.show()
```

```python
def linear_regression(x, model):
    return model.predict([[x]])[0]

new_x = 0
prediction = linear_regression(new_x, model)
print(f"Prediction for x={new_x}: {prediction}")

new_x = 6
prediction = linear_regression(new_x, model)
print(f"Prediction for x={new_x}: {prediction}")
```
